The dataset used for the project has 680 accounting metrics for 3,365 Canadian firms for years 2009 to 2018 collected quarterly (4 rows for every firm per year). Hence, total rows in the dataset were 10 x 3365 x 4 = 134600. After some essential feature selection and data cleaning, I received data for 10 accounting metrics that are comprehensive enough to capture much of the firm's behaviour. K-Means clustering is performed on every combination of two on these 10 accounting constructs to identify the which constructs are capable of revealing better clusters, i.e. more similarity.

Importing the essential libraries. For this pairwise clustering, K is set to "2". The clusters are quantified using Silhouette Co-efficient (SC). Clusters are said to be good when intra-cluster distance is minimized and inter-cluster distance is minimized. For any given clusters, the value of SC ranges from 0 to 1, where 1 indicates perfect clusters and a value closer to 0 means poor clusters or clusters are not self identifiable. SC value is based on intra and inter cluster distances.

```python
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#df = pd.read_csv("CA2.csv")
main = pd.read_csv("CA.csv")
```

```python
#df1 = df.fillna(df.mean())
main = main.fillna(main.mean())
```

Nested for loop to created 45 cluster graphs with thier SC values. This analysis helps us to further differentiate these accounting constructs. Here, using a separate tabular analysis it can be seen that retained earnings, revenue, EPS and net assets are most significant.

Due to above analysis, cluster analysis using these 4 (retained earning, revenue, EPS, net assets) and 3 (retained earnings, revenue, EPS) was performed for significant results. The jupyter notebook of 3 and 4 elemental cluster analysis is present in this repo as "Cluster_Analysis_2.ipynb"

```python
l = []
list_ = main.columns
for i in list_:
    if i != "GVKEY":
        l.append(i)

c=0

# print(l[5])
for i in range(10):
    for j in range(i+1,10):
        temp = pd.merge(main[l[i]],main[l[j]], on=main["GVKEY"])
        temp = temp.iloc[:,1:]
        model = KMeans(n_clusters=2, random_state=1).fit(temp)
        s_coeff = silhouette_score(temp, model.labels_)
        colors = ['yellow', 'cyan']
        print("Silhouette coefficient of clusters " + str(l[i]) + " and " + str(l[j]) + " is " + str(s_coeff))
        c=c+1
        temp=temp.to_numpy()
        plt.figure()
        for dot, label in zip(temp, model.labels_):
            plt.scatter(dot[0],dot[1], c=colors[label])
        for center in model.cluster_centers_:
            plt.scatter(center[0], center[1], c="black", marker="^")
        plt.show()
#         print(str(i) + " " + str(j))
```
